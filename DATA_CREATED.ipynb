## Packages

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Graph with six groups

```python
G = nx.random_partition_graph([60,60,60,60,60,60,60], 0.2, 0.001, seed = 42)
def colorear_grafo(grafo):
    pos = nx.spring_layout(grafo, seed=6)
    nx.draw(grafo, pos=pos)
    plt.show()
colorear_grafo(G)
```

## Triangular Lattice

```python
G = nx.triangular_lattice_graph(10,10)
mapeo_nombres_a_numeros = {nodo: i for i, nodo in enumerate(G.nodes)}
G2 = nx.relabel_nodes(G, mapeo_nombres_a_numeros, copy=True)

def colorear_grafo(grafo, enlaces_colores):
    pos = nx.spring_layout(grafo, seed=6)
    nx.draw(grafo, pos=pos, edge_color=[enlaces_colores[enlace] for enlace in grafo.edges])
    plt.show()

margen_inferior = 80
margen_superior = 128
lista_enlaces_importantes = list(G2.edges())[margen_inferior:margen_superior]
lista_enlaces_no_importantes = list(G2.edges())[:margen_inferior] + list(G2.edges())[margen_superior:]
enlaces_colores1 = {i:"red" for i in lista_enlaces_importantes}
enlaces_colores2 = {i:"blue" for i in lista_enlaces_no_importantes}
enlaces_colores = {**enlaces_colores1, **enlaces_colores2}
colorear_grafo(G2, enlaces_colores)
```

## Moore Neighbour

```python
def moore_neighbours_graph(width, height):
    G = nx.grid_2d_graph(width, height)
    for node in G.nodes:
        x, y = node
        neighbors = [(x-1, y-1), (x-1, y), (x-1, y+1),
                     (x, y-1),             (x, y+1),
                     (x+1, y-1), (x+1, y), (x+1, y+1)]
        for neighbor in neighbors:
            if neighbor in G.nodes:
                G.add_edge(node, neighbor)
    return G
width = 10
height = 10
G = moore_neighbours_graph(width, height)
```

```python
mapeo_nombres_a_numeros = {nodo: i for i, nodo in enumerate(G.nodes)}
G2 = nx.relabel_nodes(G, mapeo_nombres_a_numeros, copy=True)
def colorear_grafo(grafo, enlaces_colores):
    pos = nx.spring_layout(grafo, seed=6)
    nx.draw(grafo, pos=pos, edge_color=[enlaces_colores[enlace] for enlace in grafo.edges])
    plt.show()
margen_inferior = 200
margen_superior = 280
lista_enlaces_importantes = list(G2.edges())[margen_inferior:margen_superior]
lista_enlaces_no_importantes = list(G2.edges())[:margen_inferior] + list(G2.edges())[margen_superior:]
enlaces_colores1 = {i:"red" for i in lista_enlaces_importantes}
enlaces_colores2 = {i:"blue" for i in lista_enlaces_no_importantes}
enlaces_colores = {**enlaces_colores1, **enlaces_colores2}
colorear_grafo(G2, enlaces_colores)
```
